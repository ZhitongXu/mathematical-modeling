```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
dat = pd.read_csv('hairdryer_cleaned.csv', header=0)
X = dat['review_body']
```

```python
# sentiment score   
res = []

sid = SentimentIntensityAnalyzer()
for sen in X:
    ss = sid.polarity_scores(sen)
    tmp = []
    for k in ss:
        tmp.append(ss[k])
    res.append(tmp)

labels = ['neg', 'neu', 'pos', 'compound']
df = pd.DataFrame.from_records(res, columns=labels)
```
